```python
%%R

# smoothline assuming that the petal-width(in y-axis) is dependent on the sepal-width(in x-axis) with customized line method
n4 <- ggplot(data=iris, aes(y=Petal.Width, x=Sepal.Width, col=Species))+geom_smooth(method=lm, se=F)+geom_point()
```
